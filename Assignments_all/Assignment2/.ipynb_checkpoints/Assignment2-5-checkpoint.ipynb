```python
import torch.nn as nn
import torch
from PIL import Image
import numpy as np
import torch.optim as optim
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from torch.autograd import Variable
import os
```

```python
# Get the gesture dataset and font1 dataset
gesture_dataset = np.zeros([24, 2304])
font1_dataset = np.zeros([24, 256])

for i in range(65, 90):
    if i == 74:
        continue
    gesture_img_dir = './gesture_input/' + chr(i) + 'h.png'
    
    font1_img_dir = './font1_input/' + chr(i) + '1.png'
    
    gesture_img = Image.open(gesture_img_dir)
    gesture_img = gesture_img.convert("L")
    gesture_img.save(gesture_img_dir)
    
    font1_img = Image.open(font1_img_dir)
    font1_img = font1_img.convert("1")
    font1_img.save(font1_img_dir)
    
    gesture_data = gesture_img.getdata()
    gesture_array = np.array(gesture_data)/255
    
    font1_data = font1_img.getdata()
    font1_array = np.array(font1_data)/255
    
    if i < 74:
        gesture_dataset[i - 65] = gesture_array
        font1_dataset[i - 65] = font1_array
    else :
        gesture_dataset[i - 66] = gesture_array
        font1_dataset[i - 66] = font1_array
print("gesture: " , gesture_dataset)
print("font1: " , font1_dataset)
```

```python
# Define a DNN
class DNN(nn.Module):
    
    def __init__(self, input_size, hidden1_size, hidden2_size, hidden3_size, num_classes):
        super(DNN, self).__init__()
        self.inputlayer = nn.Linear(input_size, hidden1_size)
        self.hiddenlayer1 = nn.Linear(hidden1_size, hidden2_size)
        self.hiddenlayer2 = nn.Linear(hidden2_size, hidden3_size)
        self.outputlayer = nn.Linear(hidden3_size, num_classes)
        self.activate = nn.ReLU()

    def forward(self, x):
        res = self.inputlayer(x)
        res = self.activate(res)
        res = self.hiddenlayer1(res)
        res = self.activate(res)
        res = self.hiddenlayer2(res)
        res = self.activate(res)
        res = self.outputlayer(res)
        res = self.activate(res)
        return res
```

```python
class GestureDataset(Dataset):
    def __init__(self, dataset, label_list):
        self.dataset = dataset
        self.label_list = label_list
    def __len__(self):
        return len(self.label_list)
    def __getitem__(self, idx):
        data = self.dataset[idx]
        label = self.label_list[idx]
        return {
            'data': torch.from_numpy(data).float(),
            'label': torch.from_numpy(label).float()
        }
```

```python
# Parameters of training
input_size = 2304
num_classes = 256

hidden1_size = 1024
hidden2_size = 512
hidden3_size = 256

learning_rate = 0.001
batch_size = 24
num_epochs = 800

# Load the dataset
train_dataset = GestureDataset(dataset = gesture_dataset, label_list = font1_dataset)
train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=False)
device = torch.device('cpu')
model = DNN(input_size=input_size, hidden1_size = hidden1_size, hidden2_size = hidden2_size, hidden3_size = hidden3_size, num_classes=num_classes).to(device)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
def train(dataloader, model, num_epochs):
    for epoch in range(num_epochs):
        losses = []
        ERROR_Train = [] 
        model.train() 
        for i, data in enumerate(dataloader, 0):
            model.zero_grad()
            real_cpu, label_cpu = data['data'], data['label']
#             if torch.cuda.is_available():
#                 real_cpu = real_cpu.cuda() 
#                 label_cpu = label_cpu.cuda()
            real = real_cpu
            label = label_cpu
            inputv = Variable(real)
            labelv = Variable(label)
            output = model(inputv)
            err = criterion(output, labelv) 
            err.backward() 
            optimizer.step() 

            losses.append(err.data.item())
            error = label - output.data
#             print(error.shape)
            ERROR_Train.extend(error)
#         print(ERROR_Train)
        MAE = torch.mean(torch.abs(torch.stack(ERROR_Train)))
        ME = torch.mean(torch.stack(ERROR_Train))
        STD = torch.std(torch.stack(ERROR_Train)) 
        if epoch % 10 == 0 or epoch == num_epochs - 1:
            print('[%d/%d] Loss: %.4f MAE: %.4f Mean Error: %.4f STD: %.4f' % (epoch, num_epochs, np.average(losses), MAE, ME, STD))
    return output, model
```

```python
output, model = train(train_loader, model, num_epochs)
```

```python
if not os.path.exists('./output'):
    os.mkdir('./output')
output_np = output.detach().numpy()
output_dataset = np.zeros([24, 256])
for i in range(65, 90):
    if i == 74:
        continue
    if i < 74:
        index = i - 65
    else :
        index = i - 66
    output_img = output_np[index].reshape(16, 16)*255
    img = Image.fromarray(np.uint8(output_img))
#     img = img.convert("1")
    threshold = 50
    img = img.point(lambda p: p > threshold and 255)
    output_path = './output/' + chr(i) + '.png'
    img.save(output_path)
    data = img.getdata()
    array = np.array(data)/255
    output_dataset[index] = array
```

```python
# Calculate Fh
def calculateFh(label_dataset, output_dataset):
    x, y = label_dataset.shape
    Fh_denominator = 0    # Fh分母
    Fh_numerator = 0      # Fh分子
    Fh_array = np.zeros([x])
    for j in range(x):
        for i in range(y):
            if label_dataset[j][i] == 0:
                Fh_denominator = Fh_denominator + 1
                if output_dataset[j][i] == 0:
                    Fh_numerator = Fh_numerator + 1
        Fh = Fh_numerator / Fh_denominator
        Fh_array[j] = Fh
    return Fh_array

# Calculate Ffa
def calculateFfa(label_dataset, output_dataset):
    x, y = label_dataset.shape
    Ffa_denominator = 0    # Ffa分母
    Ffa_numerator = 0      # Ffa分子
    Ffa_array = np.zeros([x])
    for j in range(x):
        for i in range(y):
            if label_dataset[j][i] == 1:
                Ffa_denominator = Ffa_denominator + 1
            if output_dataset[j][i] == 0 and label_dataset[j][i] == 1:
                Ffa_numerator = Ffa_numerator + 1
        Ffa = Ffa_numerator / Ffa_denominator
        Ffa_array[j] = Ffa
    return Ffa_array
```

```python
Fh_array = calculateFh(font1_dataset, output_dataset)
Ffa_array = calculateFfa(font1_dataset, output_dataset)
print(Fh_array)
print(Ffa_array)
```

```python
def gaussian_noise(img, mean, sigma, percent):
    # Generate gauss noise
    #print(img)
    gaussian_out=img.copy()
    noise = np.random.normal(mean, sigma, int(len(img) * percent))
    #print("this is noise:",noise)
    # Add the noise to image
    gp = np.random.uniform(0,2304,int(len(img) * percent))
    for i in range(int(len(img) * percent)):
        if gaussian_out[int(gp[i])]==1:
            gaussian_out[int(gp[i])]-=abs(noise[i])
        else:
            gaussian_out[int(gp[i])]+=abs(noise[i])
    #gaussian_out = img + noise
    # Make the value between 0 and 1
    gaussian_out = np.clip(gaussian_out, 0, 1)
    return gaussian_out
```

```python
std = [0.001, 0.002, 0.003, 0.005, 0.01, 0.02, 0.03, 0.05, 0.1]
percent = [0.1, 0.2, 0.25, 0.3, 0.35]

gaussian_dataset = np.zeros([len(percent), len(std), 24, 2304])

if not os.path.exists('./input_noise/'):
    os.mkdir('./input_noise/')

for i in range(len(percent)):
    if not os.path.exists('./input_noise/' + str(percent[i]*100) + '%'):
        os.mkdir('./input_noise/' + str(percent[i]*100) + '%')
    for j in range(len(std)):
        if not os.path.exists('./input_noise/' + str(percent[i]*100) + '%/' + str(std[j])):
            os.mkdir('./input_noise/' + str(percent[i]*100) + '%/' + str(std[j]))
        for k in range(65, 90):
            if k == 74:
                continue
            if k < 74:
                index = k - 65
            else :
                index = k - 66
            inputImage = gesture_dataset[index]
            gaussian_data = gaussian_noise(inputImage, 0, std[j], percent[i])
            img = gaussian_data.reshape(48, 48)*255
            img = Image.fromarray(np.uint8(img))
            img.convert("L")
            inputImageDir = './input_noise/' + str(percent[i]*100) + '%/' + str(std[j]) + '/' + chr(k) + '.png'
            img.save(inputImageDir)
            
            gaussian_dataset[i][j][index] = gaussian_data
gaussian_dataset = np.array(gaussian_dataset)

Fh_noise_array = np.zeros([len(percent), len(std), 24])
Ffa_noise_array = np.zeros([len(percent), len(std), 24])

if not os.path.exists('./output_noise/'):
    os.mkdir('./output_noise/')
for i in range(len(percent)):
    if not os.path.exists('./output_noise/' + str(percent[i]*100) + '%'):
        os.mkdir('./output_noise/' + str(percent[i]*100) + '%')
    for j in range(len(std)):
        if not os.path.exists('./output_noise/' + str(percent[i]*100) + '%/' + str(std[j])):
            os.mkdir('./output_noise/' + str(percent[i]*100) + '%/' + str(std[j]))
#         train_noise_dataset = GestureDataset(dataset = gaussian_dataset[i][j], label_list = font1_dataset)
#         train_noise_loader = DataLoader(dataset=train_noise_dataset, batch_size=batch_size, shuffle=False)
#         print('Training dataset with noise standard deviation ' + str(std[j]) + ' in ' + str(percent[i]*100) + '% pixels')
        
#         model_noise = DNN(input_size=input_size, hidden1_size = hidden1_size, hidden2_size = hidden2_size, hidden3_size = hidden3_size, num_classes=num_classes).to(device)
#         criterion = nn.MSELoss()
#         optimizer = optim.Adam(model_noise.parameters(), lr=learning_rate)
#         output_noise, model_noise = train(train_noise_loader, model_noise, num_epochs)
#         output_noise_np = output_noise.detach().numpy()
        
        
        
        output_noise = model(torch.from_numpy(gaussian_dataset[i][j]).float())
        output_noise_np = output_noise.detach().numpy()     # Get the output
        output_noise_dataset = np.zeros([24, 256])
        
        for k in range(65, 90):
            if k == 74:
                continue
            if k < 74:
                index = k - 65
            else :
                index = k - 66
            output_noise_img = output_noise_np[index].reshape(16, 16)*255
            img = Image.fromarray(np.uint8(output_noise_img))
#             img = img.convert("1")
            threshold = 50
            img = img.point(lambda p: p > threshold and 255)
            output_path = './output_noise/' + str(percent[i]*100) + '%/' + str(std[j]) + '/' + chr(k) + '.png'
            img.save(output_path)
            data = img.getdata()
            array = np.array(data)/255
            output_noise_dataset[index] = array
        Fh = calculateFh(font1_dataset, output_noise_dataset)
        Ffa = calculateFfa(font1_dataset, output_noise_dataset)
        Fh_noise_array[i][j] = Fh
        Ffa_noise_array[i][j] = Ffa
print('------------Fh_noise_array------------')
print(Fh_noise_array)
print('------------Ffa_noise_array------------')
print(Ffa_noise_array)
            
            
```
